```python
import json
from collections import defaultdict
path="/home/jay/python/Data/usagov_bitly_data2012-03-16-1331923249.txt"
# records=[print(line) for line in open(path)]
records=[json.loads(line) for line in open(path)]
print(records[0]['tz'])

def get_counts(sequence):
    counts={}
    for x in sequence:
        if x in counts:
            counts[x]+=1
        else:
            counts[x]=1
    return counts

def get_counts2(sequence):
    counts=defaultdict(int)
    for x in sequence:
        counts[x]+=1
    return counts


print(get_counts(records[0]))
print(records)
```

```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
time_zones[:10]
print(len(time_zones))
```

```python
from collections import Counter
counts = Counter(time_zones)
print(counts.most_common(10))
```
